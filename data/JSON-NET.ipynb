```python
import pandas as pd
import networkx as nx
from networkx.readwrite import json_graph
from matplotlib.pylab import plt
%matplotlib inline

G = nx.Graph()
```

```python
! cd data && ls
```

```python
# Read csv for nodes and edges using pandas:
nodes = pd.read_csv("data/nodes.csv")
edges = pd.read_csv("data/edges.csv")
```

```python
# Dataframe to list:
nodes_list = nodes.values.tolist()
edges_list = edges.values.tolist()
```

```python
for i in nodes_list:
    G.add_node(i[0], name=i[1], group=i[2])
    
```

```python
for i in edges_list:
    G.add_edge(i[0],i[1], value=i[2])
```

```python
# Visualize the network:
nx.draw_networkx(G)

# Lots of nodes without any group, creates a graph as follows...
```

```python
import json
```

```python
j = json_graph.node_link_data(G)
 
js = json.dumps(j, ensure_ascii=False, indent=2)
with open("data/data.json", "w") as file:
     file.write(js)
```
